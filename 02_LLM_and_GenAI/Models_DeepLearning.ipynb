# Deep Learning Models

## 1. Siamese Network

- Siamese networks are a type of neural network architecture that contain two or more identical subnetworks. 
- The **subnetworks share the same weights** and learn to solve the same problem, **but with different inputs**. 
- The subnetworks are trained to **minimize the difference between their outputs on pairs of inputs that are similar, and maximize the difference between their outputs on pairs of inputs that are dissimilar**.
- SentenceBert is a variant of Siamese Network.

## 3. Long Short-Term Memory (LSTM)

LSTM (Long Short-Term Memory) is **a type of Recurrent Neural Network (RNN)** that is widely used in Natural Language Processing (NLP) and speech recognition tasks. It is designed to **handle the issue of vanishing gradients** in traditional RNNs by introducing a memory cell, gates (input, forget, output) and cell state that control the flow of information into and out of the memory cell. Due to the additive update mechanism, the LSTM's memory cell ensures gradients remain consistent over lengthy sequences.

The basic idea behind LSTMs is to allow information to persist in the network for a longer time period. The input gate regulates the flow of input data into the memory cell. The forget gate decides what information should be discarded from the memory cell. The output gate controls the flow of information from the memory cell to the output. The cell state acts as a “memory” that retains information for an extended period of time.

The LSTM network takes **the current input, previous hidden state, and previous cell state as input** and **produces the current hidden state and cell state**, which are then used to make predictions. The hidden state and cell state are then passed as input to the next time step in a sequence.

Overall, LSTMs help in modeling the dependencies between elements in a sequence, making it a powerful tool in sequential data modeling.

In natural language processing (NLP), the Transformer is a neural network architecture that was introduced in a 2017 paper by Google researchers. It is used for tasks such as language translation, text summarization, and question answering.

The Transformer architecture is based on the idea of self-attention, which allows the model to weigh the importance of different parts of the input when making predictions. It uses multi-head self-attention mechanism, which means it uses multiple attention heads to weigh the importance of different parts of the input. This allows the model to better understand the relationships between words in a sentence, which helps it generate more accurate and coherent text.

The Transformer architecture also uses a technique called positional encoding, which allows the model to understand the order of words in a sentence. Additionally, it uses a feed-forward neural network to further process the output of the self-attention mechanism.

The transformer architecture has become popular in many NLP tasks and its variants such as BERT, GPT, GPT-2, GPT-3 etc have been used extensively in many NLP tasks and have provided state of the art results.

## 5. Pre-training

MLM is a pre-training technique used in NLP. It is used to train a language model by masking some of the words in a sentence and then training the model to predict the masked words.

## 6. BERT

BERT, which stands for **Bidirectional Encoder Representations from Transformers**, is based on Transformers and was introduced in a 2018 paper by Google researchers. It is used for tasks such as question answering, text classification, and named entity recognition.

The BERT framework was pre-trained using text from Wikipedia and can be fine-tuned with question and answer datasets, such as SQuAD, to generate answers to questions. It can also be fine-tuned with text classification datasets, such as the GLUE benchmark, to classify text into different categories.

BERT Base Model has 12 Layers and 110M parameters with 768 Hidden and equal embedding layers. This large size makes it very computationally heavy to train.

### BERT Input

1. BERT takes two inputs: the tokenized sentence and the segment IDs. The tokenized sentence is a list of integers, where each integer represents a token in the sentence. The segment IDs are also a list of integers, where each integer represents the segment to which the corresponding token belongs. The segment IDs are used to differentiate between the two sentences in a pair of sentences. The segment IDs are always 0 for single sentences, and 0 and 1 for pairs of sentences.

### BERT Family

1. ALBERT: A Lite BERT has 12 million parameters with 768 hidden layers and 128 embedding layers, the following 2 techniques are used
   1. Cross-layer parameter sharing: In this method, the parameter of only the first encoder is learnt and the same is used across all encoders.
   2. Factorized embedding layer parameterization: Instead of keeping the embedding layer at 768, the embedding layer is reduced by factorization to 128 layers.
2. RoBERTa: RoBERTa stands for “Robustly Optimized BERT pre-training Approach”. In many ways this is a better version of the BERT model. The key points of difference are as follows:
   1. Dynamic Masking: BERT uses static masking i.e. the same part of the sentence is masked in each Epoch. In contrast, RoBERTa uses dynamic masking, wherein for different Epochs different part of the sentences are masked. This makes the model more robust.
   2. Remove NSP Task: It was observed that the NSP task is not very useful for pre-training the BERT model. Therefore, the RoBERTa only with the MLM task.
   3. More data Points: BERT is pre-trained on “Toronto BookCorpus” and “English Wikipedia datasets” i.e. as a total of 16 GB of data. In contrast, in addition to these two datasets, RoBERTa was also trained on other datasets like CC-News (Common Crawl-News), Open WebText etc. The total size of these datasets is around 160 GB.
   4. Large Batch size: To improve on the speed and performance of the model, RoBERTa used a batch size of 8,000 with 300,000 steps. In comparison, BERT uses a batch size of 256 with 1 million steps.
3. ELECTRA: ELECTRA stands for “Efficiently Learning an Encoder that Classifies Token Replacements Accurately”. The model uses a generator-discriminator structure. Other than being the lighter version of BERT, ELECTRA has the following distinguishing features:
   1. Replaced Token Detection: Instead of MLM for pre-training, ELECTRA uses a task called “Replaced Token Detection” (RTD). In RTD, instead of masking the token, the token is replaced by a wrong token and the model is expected to classify, whether the tokens are replaced with wrong or not.
   2. No NSP pre-training is performed.

## 7. T5

T5 is a text-to-text transformer model that was trained on a large corpus of text-to-text data. T5 is a successor to BERT and GPT-2, and is the first text-to-text transformer model that is trained on a large scale with a denoising objective. The changes compared to BERT include:

1. Adding a causal decoder to the bidirectional architecture.
2. Replacing the fill-in-the-blank cloze task with a mix of alternative pre-training tasks.

## 8. ChatGPT

ChatGPT is a variant of the GPT (Generative Pre-trained Transformer) model, which is a type of language model that uses deep learning to generate human-like text. The model is trained on a large dataset of text and learns patterns in the data to generate new text that is similar to the input it was trained on.

When given a prompt or input, ChatGPT uses the patterns it learned during training to generate a response. The model uses a technique called "auto-regression," where it predicts the next word in a sentence based on the previous words. This allows it to generate text that is contextually appropriate and coherent.

It uses transformer architecture and pre-training the model on a large dataset of text, then fine-tuning it on a smaller dataset specific to the task at hand. The pre-training allows the model to learn general patterns in language, while fine-tuning allows it to focus on the specific task and generate more accurate and relevant responses.

## 13. Key Architecture Trends (2024-2026)

The LLM landscape has converged on several architectural patterns that have proven to be particularly effective:

| Trend | Description | Key Models | Benefits |
|-------|-------------|------------|----------|
| **Decoder-Only Dominance** | Encoder-decoder models (like T5) have been largely replaced by decoder-only architectures | GPT series, Llama, Mistral, Claude | Simpler architecture, better scaling, unified pre-training objective (next-token prediction) |
| **RoPE Positional Encoding** | Rotary Position Embeddings replacing absolute/sinusoidal encodings | Llama, Mistral, PaLM | Better length extrapolation, relative position awareness, no learned parameters |
| **GQA/MQA Attention** | Grouped Query Attention (GQA) and Multi-Query Attention (MQA) replacing standard Multi-Head Attention | Llama 2/3, GPT-4 variants | Reduced KV cache size (up to 8x), faster inference, minimal quality loss |
| **FlashAttention** | Algorithm-level optimization for memory-efficient attention computation | Used in training most modern LLMs | Enables longer sequences, faster training, reduced memory footprint |
| **Mixture of Experts (MoE)** | Sparse models with multiple expert networks, only subset activated per token | GPT-4, Mixtral, Gemini | Scaling parameters without proportional compute increase, better efficiency |
| **KV Cache Optimization** | Techniques to reduce memory during inference (quantization, eviction, compression) | All production LLMs | Enables longer context, faster generation, lower deployment cost |
| **SwiGLU/GeGLU Activations** | Gated activation functions replacing ReLU in FFN layers | Llama, PaLM | Better training dynamics, improved performance |
| **RMSNorm over LayerNorm** | Simplified normalization that removes mean-centering | Llama, Mistral | Faster computation, similar or better performance |
| **Larger Vocabularies** | Expanding tokenizer vocabulary (128K+ tokens) | Llama 3, Gemini | Better multilingual support, reduced token count, improved efficiency |
| **Extended Context Windows** | Moving from 2K→4K→8K→32K→100K+ tokens | Claude (200K-1M), GPT-4 (32K-128K), Gemini (1M+) | Handle longer documents, better few-shot learning, improved reasoning |

### Complexity Analysis

**Standard Attention**: O(n² × d) time, O(n² + n × d) memory  
**FlashAttention**: O(n² × d) time, O(n × d) memory  
**Sparse/Local Attention**: O(n × w × d) time where w << n  
**MoE Forward Pass**: O(k/n × total_params) where k is active params

### The Inference Efficiency Stack

Modern production LLMs combine multiple techniques:
1. **Model Architecture**: GQA/MQA, RoPE, efficient activations
2. **Attention Optimization**: FlashAttention, sliding windows, sparse patterns
3. **Quantization**: INT8/INT4 weights, reduced precision
4. **KV Cache Management**: Compression, eviction policies
5. **Speculative Decoding**: Small model drafts, large model verifies
6. **Continuous Batching**: Dynamic batching for better throughput

## 12. Claude Architecture Insights

Claude, developed by Anthropic, represents a different approach to LLM development with a strong focus on safety and helpfulness.

### Constitutional AI (CAI)

Claude's training incorporates Constitutional AI, a novel approach to alignment:

1. **Supervised Learning Phase**: Model is trained on human-written examples
2. **Self-Critique Phase**: 
   - Model generates responses to prompts
   - Model critiques its own responses based on a "constitution" (set of principles)
   - Model revises responses to better align with principles
3. **RLAIF (Reinforcement Learning from AI Feedback)**:
   - Instead of RLHF using human feedback, uses AI feedback from the self-critique process
   - Scales more efficiently than pure RLHF while maintaining alignment

### Long Context Processing

Claude pioneered ultra-long context windows:
- Claude 2: 100K tokens
- Claude 2.1: 200K tokens  
- Claude 3: 200K tokens with experimental 1M+ token support

**Technical Approaches**:
- Advanced attention optimization techniques
- Efficient KV cache management
- Maintains quality even at extreme context lengths (needle-in-haystack tests)
- Uses positional encoding schemes that extrapolate well to long sequences

### Preference-Based Training

- Heavy emphasis on being helpful, harmless, and honest (HHH criteria)
- Multi-stage RLHF/RLAIF focusing on:
  - Reducing harmful content
  - Improving factual accuracy
  - Better instruction following
  - Appropriate refusals (saying no to harmful requests)

### Architecture Details

While exact architecture is not public, Claude is known to use:
- Decoder-only transformer architecture
- Modern attention mechanisms (likely GQA or similar)
- Advanced positional encoding for long context
- Efficient inference optimizations

## 11. Mistral and Mixtral

Mistral AI has developed efficient open-source models that punch above their weight class through architectural innovations.

### Mistral 7B

**Key Innovation: Sliding Window Attention (SWA)**
- Instead of full attention over all previous tokens, each token only attends to the previous W tokens (window size)
- Significantly reduces computation from O(n²) to O(n×W)
- Despite local attention, information can propagate across the entire sequence through multiple layers
- Achieves competitive performance with much larger models while being faster and more memory-efficient
- Native 8K context length with efficient processing

### Mixtral 8x7B

**Sparse Mixture of Experts (SMoE) Architecture**
- Contains 8 expert networks (each 7B parameters)
- For each token, a router network selects the top-2 experts to process it
- Total parameters: 47B, but only ~13B active per token
- Benefits:
  - Quality of a 47B model with inference cost of a 13B model
  - Different experts specialize in different types of inputs (languages, domains, reasoning patterns)
  - Better scaling efficiency compared to dense models

**Sparse Gating Mechanism**
- Router uses a learned gating function: `G(x) = Softmax(TopK(x · W_g))`
- Top-K selection ensures sparsity (only K experts activated)
- Load balancing loss ensures experts are utilized evenly during training

### Mixtral 8x22B

- Scales up to 22B parameters per expert (176B total, ~44B active)
- Maintains the efficient sparse activation pattern
- Extended context window support

## 10. Llama Family

Meta's Llama (Large Language Model Meta AI) series represents a major contribution to open-source LLMs, with architectural innovations that have influenced the broader field.

### Architecture Evolution

**Llama 1 (2023)**
- 7B, 13B, 33B, and 65B parameter models
- Decoder-only transformer architecture
- Uses RMSNorm instead of LayerNorm for better training stability
- Introduced RoPE (Rotary Position Embeddings) for positional encoding

**Llama 2 (2023)**
- 7B, 13B, and 70B parameter models
- Trained on 2 trillion tokens (40% more than Llama 1)
- Introduced Grouped Query Attention (GQA) for efficient inference
- 4096 context length with optimized attention

**Llama 3 (2024)**
- 8B and 70B parameter models (also 405B)
- Trained on 15 trillion tokens
- Extended vocabulary size (128K tokens) for better multilingual support
- 8K native context length with 128K extended variants

### Key Architectural Features

1. **RoPE (Rotary Position Embeddings)**: Instead of adding positional information to embeddings, RoPE rotates the query and key vectors based on position. This provides better extrapolation to longer sequences and relative position awareness.

2. **Grouped Query Attention (GQA)**: A hybrid between Multi-Head Attention (MHA) and Multi-Query Attention (MQA). In GQA, queries are divided into groups, and each group shares a single key-value head. This reduces the KV cache size while maintaining model quality, making inference more memory-efficient.

3. **SwiGLU Activation**: Uses SwiGLU (Swish-Gated Linear Unit) instead of ReLU in the feed-forward layers for better performance.

4. **Pre-normalization**: Applies RMSNorm before each transformer sub-layer (instead of after) for more stable training.

## 9. GPT-4 and Beyond

GPT-4 represents a significant leap in large language model capabilities, introducing several architectural and scaling innovations:

### Key Features

1. **Mixture of Experts (MoE) Architecture**: GPT-4 is believed to use a sparse MoE architecture where different expert networks are activated for different inputs. This allows the model to have a large total parameter count while only activating a subset during inference, improving efficiency.

2. **Multi-modal Capabilities**: Unlike GPT-3, GPT-4 can process both text and images as input, making it a true multi-modal model. It uses vision encoders to process images and integrates visual and textual information through cross-attention mechanisms.

3. **Longer Context Windows**: GPT-4 supports significantly longer context windows (up to 32K tokens in extended versions, with experimental 128K variants), allowing it to maintain coherence over much longer conversations and documents. This is achieved through optimized attention mechanisms and improved positional encodings.

4. **Improved Reasoning and Instruction Following**: GPT-4 shows substantial improvements in complex reasoning tasks, mathematical problem solving, and following detailed instructions, likely due to enhanced RLHF (Reinforcement Learning from Human Feedback) training and better base model capabilities.

5. **Safety and Alignment**: Incorporates more sophisticated safety measures and alignment techniques to reduce harmful outputs and improve factual accuracy.
